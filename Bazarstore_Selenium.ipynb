```python
import pandas as pd
import numpy as np
import selenium
from selenium import *
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time 
import os
import re
pd.set_option('display.max_rows', None)
```

```python
PATH = "/Users/FaiqRasulov/Desktop/chromedriver"
driver = webdriver.Chrome(PATH)
MAX_PAGE_NUM =11
MAX_PAGE_DIG = 1

with open('results.csv', 'w') as f:
    f.write("Name, Price \n")
    cont_list =[]
for i in range(1, MAX_PAGE_NUM + 1):
    page_num = (MAX_PAGE_DIG - len(str(i))) * "0" + str(i)
    url = "https://bazarstore.az/16-sued-shr-yemyi?page=" + page_num 

    driver.get(url)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    #WebDriverWait(driver,30).until(EC.presence_of_all_elements_located((By.CLASS_NAME,"row.bs-product-row.justify-content-center")))
    container= driver.find_elements(By.CLASS_NAME, 'row.bs-product-row.justify-content-center')
    for containers in container:
        WebDriverWait(driver,5).until(EC.presence_of_all_elements_located((By.CLASS_NAME,"product-title")))
        name = containers.find_element(By.CLASS_NAME, "product-title").get_attribute("textContent").strip()
        WebDriverWait(driver,5).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR,"a[href*='https://bazarstore.az/']")))
        link = containers.find_element(By.CSS_SELECTOR, "a[href*='https://bazarstore.az/']").get_attribute("href").strip()
        WebDriverWait(driver,5).until(EC.presence_of_all_elements_located((By.CLASS_NAME,"pp_price_with_text")))
        price= containers.find_element(By.CLASS_NAME, "pp_price_with_text").get_attribute("textContent").strip()
        WebDriverWait(driver,5).until(EC.presence_of_all_elements_located((By.TAG_NAME,"span")))
        qiymet_ferq = containers.find_element(By.TAG_NAME, 'span').get_attribute("textContent").strip()

        con_item= {
        'name' : name,
        'price' : price,
        'link' : link,
        'qiymet_ferq':qiymet_ferq
        }
        cont_list.append(con_item)

df=pd.DataFrame(cont_list)
print(df)
driver.quit()
```
