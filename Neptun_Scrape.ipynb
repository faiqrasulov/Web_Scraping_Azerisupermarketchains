```python
import pandas as pd
import numpy as np
import selenium
from selenium import *
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time 
import os
import re
pd.set_option('display.max_rows', None)
```

```python
PATH = "/Users/FaiqRasulov/Desktop/chromedriver"
driver = webdriver.Chrome(PATH)
MAX_PAGE_NUM =82
MAX_PAGE_DIG = 1

with open('results.csv', 'w') as f:
    f.write("Name, Price \n")
    cont_list =[]
for i in range(1, MAX_PAGE_NUM + 1):
    page_num = (MAX_PAGE_DIG - len(str(i))) * "0" + str(i)
    url = "https://neptun.az/sud-mehsullari?page=" + page_num 
    driver.get(url)
    for j in range(1, 10):
        driver.execute_script("window.scrollTo(0, "+str(j*1000)+");")
        time.sleep(0.5)
    container= driver.find_elements(By.CLASS_NAME, 'product-item-container')
    for containers in container:
        WebDriverWait(driver,60).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR,"div.caption")))
        name = containers.find_element(By.CSS_SELECTOR, "div.caption").get_attribute("textContent").strip()
        WebDriverWait(driver,60).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR,"div.price")))
        price= containers.find_element(By.CSS_SELECTOR, "div.price").get_attribute("textContent").strip()

        con_item= {
        'name' : name,
        'price' : price,
        }
        cont_list.append(con_item)

df=pd.DataFrame(cont_list)
print(df)
driver.quit()
```

```python
import pandas as pd

# Check if the 'price' column contains discount information
if df1['price'].str.contains(' \n\n').any():
    # Split the 'price' column into 'ref_qiymet' and 'endirimsiz_qiymet' columns
    df1[['ref_qiymet', 'endirimsiz_qiymet']] = df1['price'].str.split(' \n\n', expand=True)
    
    # Extract only the digits and decimal points from the 'ref_qiymet' column
    df1['ref_qiymet'] = df1['ref_qiymet'].str.extract('([\d.]+)').astype(float)
    
    # Extract only the digits and decimal points from the 'endirimsiz_qiymet' column
    df1['endirimsiz_qiymet'] = df1['endirimsiz_qiymet'].str.extract('([\d.]+)').astype(float)
    
    # Calculate the 'endirim_faiz' column
    df1['endirim_faiz'] = ((df1['endirimsiz_qiymet'] - df1['ref_qiymet']) / df1['endirimsiz_qiymet']) * 100
    
    # Round the 'endirim_faiz' column to 2 decimal places
    df1['endirim_faiz'] = df1['endirim_faiz'].round(2)
    
    # Rearrange the columns
    df1 = df1[['tarix','brand', 'kateqoriya', 'qramaj','yag','ref_qiymet', 'endirimsiz_qiymet', 'endirim_faiz' ,'mehsul_adi']]
else:
    # Extract only the digits and decimal points from the 'price' column
    df1['ref_qiymet'] = df1['price'].str.extract('([\d.]+)').astype(float)
    
    # If there is no discount, continue with the original DataFrame
    df1 = df1[['tarix', 'brand', 'kateqoriya', 'qramaj', 'yag', 'ref_qiymet', 'mehsul_adi']]


# Rename columns by stripping leading/trailing whitespace
df1 = df1.rename(columns=lambda x: x.strip())

# Print the modified DataFrame
print(df1)
```
