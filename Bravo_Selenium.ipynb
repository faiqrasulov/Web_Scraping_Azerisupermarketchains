```python
import pandas as pd
import numpy as np
import selenium
from selenium import *
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time 
import os
import re
pd.set_option('display.max_rows', None)
```

```python
PATH = "/Users/FaiqRasulov/Desktop/chromedriver"
driver = webdriver.Chrome(PATH)
MAX_PAGE_NUM = 25
MAX_PAGE_DIG = 1
driver = webdriver.Chrome(PATH)

with open('results.csv', 'w') as f:
    f.write("Name, Price \n")
    cont_list =[]
for i in range(1, MAX_PAGE_NUM + 1):
    page_num = (MAX_PAGE_DIG - len(str(i))) * "0" + str(i)
    url = "https://bravomarket.online/catalog/dairy-and-eggs/?PAGEN_1=" + page_num + "&SIZEN_1=12l"
    #time.sleep(3)
    driver.get(url)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    #time.sleep(3)
    WebDriverWait(driver,60).until(EC.presence_of_all_elements_located((By.CLASS_NAME,"product_item.product-item")))
    container= driver.find_elements(By.CLASS_NAME, 'product_item.product-item')
    for containers in container:
        WebDriverWait(driver,60).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR,"div.name")))
        #time.sleep(3)
        name = containers.find_element(By.CSS_SELECTOR, "div.name").get_attribute("textContent").strip()
        WebDriverWait(driver,60).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR,"div.price.product-item-price-current")))
        #time.sleep(3)
        price= containers.find_element(By.CSS_SELECTOR, "div.price.product-item-price-current").get_attribute("textContent").strip()
        WebDriverWait(driver,60).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR,"div.old_price.product-item-price-old")))
        #time.sleep(3)
        endirimsiz_qiymet= containers.find_element(By.CSS_SELECTOR, "div.old_price.product-item-price-old").get_attribute("textContent").strip()
        WebDriverWait(driver,60).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR,"div.product-item-label-ring.product-item-label-small.product-item-label-bottom.product-item-label-right")))
        #time.sleep(3)
        endirim_faiz= containers.find_element(By.CSS_SELECTOR, "div.product-item-label-ring.product-item-label-small.product-item-label-bottom.product-item-label-right").get_attribute("textContent").strip()
        con_item= {
        'name' : name,
        'price' : price,
        'endirimsiz_qiymet' : endirimsiz_qiymet,
        'endirim_faiz' : endirim_faiz,
        }
        cont_list.append(con_item)

df=pd.DataFrame(cont_list)
print(df)
driver.quit()
```
